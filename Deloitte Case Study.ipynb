```python
from datetime import datetime
```

```python
# Now we need to format the Year column as YYYY-MM-DD__ Format in monthly dataframe .
Date1 = [ ]
for i in range(len(Exchange_Rate_monthly)):
    Date1.append(Exchange_Rate_monthly['Year'][i][0:4])    

# For MM format we need to selected 5th and 6th index as they represent the month number.  
Date2 = [ ]
for i in range(len(Exchange_Rate_monthly)):
    Date2.append(Exchange_Rate_monthly['Year'][i][5:7])    

# Now to add YYYY and MM together we need to create another list
Date3 = [ ]
for i in range(len(Date1)):
    Date3.append(Date2[i]+'-'+Date1[i])
    
def prepend(list, str):
 
    # Using format()
    str += '{0}'
    list = [str.format(i) for i in list]
    return(list)
a = '01-'
Date3 = prepend(Date3,a)

Date3 = [datetime.strptime(x,'%d-%m-%Y') for x in Date3]

# Find the name of the column by index
n = Exchange_Rate_monthly.columns[0]

# Drop that column
Exchange_Rate_monthly.drop(n, axis = 1, inplace = True)

# Put whatever series you want in its place
Exchange_Rate_monthly.insert(0,'Year',Date3)
```
